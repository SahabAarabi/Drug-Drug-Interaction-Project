```python
# Define input and output paths
input_path = "/content/drive/My Drive/Drug Interaction/project_results/merged_results_first_edition_without_change.csv"
output_path = "/content/drive/My Drive/Drug Interaction/project_results/merged_results_second_edition_without_mirror_interactions.csv"

# Define chunk size (adjust if needed)
chunk_size = 100000  # Process 100K rows at a time

# Define severity order for comparison
severity_order = {"Contraindicated": 3, "Major": 2, "Moderate": 1}

# Dictionary to store processed results
interaction_dict = {}  # {pre_id: set((drug1, drug2))} to track mirrored pairs
interaction_counts = {}  # {pre_id: count}

# Open output file for writing
with open(output_path, "w", newline="") as out_file:
    writer = None  # Will be initialized after reading the first chunk

    # Process file in chunks
    for chunk in pd.read_csv(input_path, chunksize=chunk_size):
        processed_rows = []

        for _, row in chunk.iterrows():
            pre_id = row["pre_id"]
            drug1, drug2 = row["prescribed_drug"], row["interacting_drug_full_name"]
            pair = tuple(sorted([drug1, drug2]))  # Standardized interaction pair
            severity = row['severity']
            # Initialize pre_id in dictionaries if not exists
            if pre_id not in interaction_dict:
                interaction_dict[pre_id] = {}
                interaction_counts[pre_id] = 0


            # Check if mirrored interaction already exists
            if pair not in interaction_dict[pre_id]:
                # Add new pair if not exists
                interaction_dict[pre_id][pair] = {"severity": severity, "row": row}
                interaction_counts[pre_id] += 1  # Increase interaction count
                processed_rows.append(row)  # Keep this row
            else:
                # Compare severity with the existing row
                existing_severity = interaction_dict[pre_id][pair]["severity"]

                if severity_order[severity] > severity_order[existing_severity]:
                    # Replace the existing row with the current row (higher severity wins)
                    interaction_dict[pre_id][pair] = {"severity": severity, "row": row}

                    # Remove the old row and add the new row to processed_rows
                    processed_rows = [r for r in processed_rows if tuple(sorted([r["prescribed_drug"], r["interacting_drug_full_name"]])) != pair]
                    processed_rows.append(row)  # Add the new higher-severity row

        # Convert processed rows into a DataFrame
        df_processed = pd.DataFrame(processed_rows)

        # Update interaction_count for each row before saving
        df_processed["interaction_count"] = df_processed["pre_id"].map(interaction_counts)

        # Write to CSV (initialize headers only once)
        if writer is None:
            writer = csv.DictWriter(out_file, fieldnames=df_processed.columns)
            writer.writeheader()  # Write header on first write

        df_processed.to_csv(out_file, index=False, mode="a", header=False)  # Append data

print(f"âœ… Processed file saved at: {output_path}")
```

```python
import pandas as pd

# Define file paths
input_path = "/content/drive/My Drive/Drug Interaction/project_results/merged_results_second_edition_without_mirror_interactions.csv"  # Input file
output_path = "/content/drive/My Drive/Drug Interaction/project_results/merged_results_second_edition_without_mirror_interactions_corrected.csv"  # Output file

# Define chunk size to avoid memory issues
chunk_size = 100000

# Step 1: Count interactions per pre_id
interaction_counts = {}

for chunk in pd.read_csv(input_path, chunksize=chunk_size):
    pre_id_counts = chunk["pre_id"].value_counts().to_dict()
    for pre_id, count in pre_id_counts.items():
        interaction_counts[pre_id] = interaction_counts.get(pre_id, 0) + count

# Step 2: Update interaction_count and save to a new file
with open(output_path, "w") as out_file:
    writer = None

    for chunk in pd.read_csv(input_path, chunksize=chunk_size):
        chunk["interaction_count"] = chunk["pre_id"].map(interaction_counts)

        # Write updated data to a new file
        if writer is None:
            chunk.to_csv(out_file, index=False)  # Write with header
            writer = True
        else:
            chunk.to_csv(out_file, index=False, header=False, mode="a")  # Append without header

print(f"âœ… Interaction count corrected and saved at: {output_path}")
```

```python
import pandas as pd
import re

# Load the CSV file
file_path = "Book2.csv"  # Update with your file path
df = pd.read_csv(file_path)

# Function to clean drug entries (remove percentages)
def clean_drug_entry(entry):
    if pd.isna(entry):  # Handle NaN values
        return None
    return re.sub(r"\s*\(\d+\.\d+%\)", "", entry)  # Remove percentages

# Apply cleaning function to all relevant columns
for col in df.columns[1:]:  # Skipping the 'name' column
    df[col] = df[col].apply(clean_drug_entry)

df
```

```python

# Function to remove "severity : *" from a column
def remove_severity(text):
    if pd.isna(text):  # Handle NaN values
        return text
    return re.sub(r"\s*-\s*Severity:\s*[^,]+", "", text)  # Remove (severity : ...)

# Apply function to 'top_5_all' column
df["top_5_all"] = df["top_5_all"].apply(remove_severity)
df
```

```python
# Function to expand rows for contraindicated agents
def expand_rows(df, col_name):
    expanded_rows = []
    for _, row in df.iterrows():
        drugs = row[col_name]
        if pd.isna(drugs):  # Skip empty entries
            expanded_rows.append(row)
        else:
            drug_list = [drug.strip() for drug in drugs.split(",")]  # Split by comma
            for drug in drug_list:
                new_row = row.copy()
                new_row[col_name] = drug  # Keep only one drug per row
                expanded_rows.append(new_row)
    return pd.DataFrame(expanded_rows)
```

```python
# Expand the contraindicated agents column
df_expanded_contraindicated = expand_rows(df, "contraindicated agents").drop(columns = ['top_moderate', 'top_major', 'top_5_all'])
df_expanded_contraindicated

# Save the processed file
processed_file_path = "contraindicated.csv"
df_expanded_contraindicated.to_csv(processed_file_path, index=False)

print("Processing complete. File saved as:", processed_file_path)
```

```python
import pandas as pd
import re
from tqdm import tqdm  # Progress bar library

# Ensure all values are strings to avoid decoding errors
contraindicated["contraindicated agents"] = contraindicated["contraindicated agents"].astype(str).str.lower()
df_prescriptions["pres_full_name"] = df_prescriptions["pres_full_name"].astype(str).str.lower()

# Function to count occurrences of an agent in pres_full_name
def count_occurrences(agent, df_column):
    return df_column.str.contains(re.escape(agent), case=False, na=False).sum()

# Get unique agents to avoid redundant calculations
unique_agents = contraindicated["contraindicated agents"].dropna().unique()

# Use tqdm for progress tracking
tqdm.pandas(desc="Processing Unique Interacting Agents")

# Create a dictionary to store results
agent_counts = {agent: count_occurrences(agent, df_prescriptions["pres_full_name"]) for agent in tqdm(unique_agents)}

# Map stored values back to the DataFrame (avoiding recalculations)
contraindicated["interacting_agent_count_all"] = contraindicated["contraindicated agents"].map(agent_counts)

print("Processing complete. Updated dataframe saved.")
```
